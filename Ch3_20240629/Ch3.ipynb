````python
df[df["total_bill"] > 30].head()
````

````python
df[df['sex'] == 'Male'][0:5]
````

````python
df[(df["total_bill"] > 30) & (df["sex"] == "Male") ].head()
````

````python
df[(df["day"] == "Sun") | (df["day"] == "Sat")]
````

````python
sex = "Female"
df.query(f'total_bill > 30 and sex == "{sex}"' )
````

````python
df.query("day=='Sun' or day=='Sat'")
````

````python
df[(df["total_bill"] > 30) & (df["sex"] != "Male") ].head()
````

````python
df[(df["total_bill"] > 30) & ~(df["sex"] == "Male") ].head()
````
